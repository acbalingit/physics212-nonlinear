```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
#Black and white - use first channel
data = mpimg.imread('Philippines.png')[:,:,0]
data = 1 - np.floor(data) ## Flip image (zeroes outside boundaries), ensure binary
x_dim, y_dim = np.shape(data)
plt.imshow(data)
```

```python
def fractal_dimension(image_array):
    x_dim, y_dim = shape(data)
    log_box_count = []
    log_box_size = []

    for box_size in range(30, 700):
        box_count = 0
        box_data = 0
        x_positions = range(x_dim)[::box_size]
        y_positions = range(y_dim)[::box_size]

        for y_boxpos in y_positions:
            for x_boxpos in x_positions:
                box_data = box_data + 1
                ## Check if bounded region is partially filled (between 0 to box-area)
                if 0 < sum(image_array[x_boxpos:(x_boxpos+box_size), y_boxpos:(y_boxpos+box_size)]) < (box_size ** 2):
                    box_count = box_count + 1

        log_box_count.append(np.log(box_count))
        log_box_size.append(np.log(1.0/box_size))

    plt.plot(log_box_size, log_box_count)
    m, x = polyfit(log_box_size, log_box_count, 1)
    return m
```

```python
fractal_dimension(data)
```

```python
X, Y = np.meshgrid(np.linspace(-1, 1, x_dim), np.linspace(-1, 1, x_dim))
circle = 1. * np.sqrt(X**2 + Y**2) < 0.6
plt.imshow(circle)
```

```python
square = (Y < 0.6) * (Y > -0.6) * (X < 0.6) * (X > -0.6)
plt.imshow(square)
```

```python
fractal_dimension(circle)
```

```python
fractal_dimension(square)
```
